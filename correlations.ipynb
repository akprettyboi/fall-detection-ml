```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Code to read your Excel file
df = pd.read_csv('C:\\Users\\nimay\\Desktop\\Fall Detection\\ImageProccesing\\HAR UP Edited Datatset.csv')

# Assuming 'df' is your DataFrame with the data from the Excel file
# and it has the same structure as the image you've provided

# Selecting the columns related to NAcc, Nvelocity, NeckLuminosity and Activity
columns_of_interest = ['NAcc-x', 'Nacc-y', 'Nacc-z', 
                       'Nvelocity-X', 'Nvelocity-Y', 'Nvelocity-z', 
                       'NeckLuminosity', 'Activity']

# Calculating the correlation matrix for the selected columns
correlation_matrix = df[columns_of_interest].corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Define the columns to include in the correlation matrix
columns_of_interest = [
    'NAcc-x', 'Nacc-y', 'Nacc-z', 'Nvelocity-X', 'Nvelocity-Y', 'Nvelocity-z', 
    'NeckLuminosity', 'Wacc-X', 'Wacc-Y', 'Wacc-z', 'Wvelocity-X', 
    'WVelocity-Y', 'Wvelocity-Z', 'WristLuminosity', 'BrainSensor', 
    'Infrared1', 'Infrared2', 'Infrared3', 'Infrared4', 'Infrared5', 'Infrared6'
]

# Create a dictionary to store each activity's correlation matrix
activity_correlation_matrices = {}

# Loop through each activity number and calculate the correlation matrix
for activity in df['Activity'].unique():
    # Filter the DataFrame for the current activity
    activity_data = df[df['Activity'] == activity][columns_of_interest]
    
    # Calculate the correlation matrix
    corr_matrix = activity_data.corr()
    
    # Store the correlation matrix in the dictionary with activity as the key
    activity_correlation_matrices[activity] = corr_matrix

# Now you have a dictionary where each key is an activity number
# and each value is the corresponding correlation matrix.
# You can access each activity's correlation matrix like this:
# activity_correlation_matrices[activity_number]

# If you want to view or do something with each correlation matrix, you can loop through the dictionary:
for activity, corr_matrix in activity_correlation_matrices.items():
    print(f"Correlation matrix for activity {activity}:")
    print(corr_matrix)
    # You can also do further analysis or plotting with each matrix here
```

```python
def plot_correlation_matrices(df, columns_of_interest):
    # Iterate over each activity number (from 1 to 11)
    for activity_num in range(1, 12):
        # Filter the DataFrame for the current activity number
        activity_df = df[df['Activity'] == activity_num]
        
        # Calculate the correlation matrix for the filtered DataFrame
        correlation_matrix = activity_df[columns_of_interest].corr()
        
        # Plotting the correlation matrix using seaborn
        plt.figure(figsize=(12, 10))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
        plt.title(f'Correlation Matrix for Activity {activity_num}')
        
        # Saving the plot as an image file
        plt.savefig(f'correlation_matrix_activity_{activity_num}.png')
        plt.close()  # Close the plot to avoid displaying it inline

# Call the function to plot and save the correlation matrix images
plot_correlation_matrices(df, columns_of_interest)
```

```python
import numpy as np

def plot_highly_positive_correlations(df, columns_of_interest, threshold=0.8):
    # Iterate over each activity number (from 1 to 11)
    for activity_num in range(1, 12):
        # Filter the DataFrame for the current activity number
        activity_df = df[df['Activity'] == activity_num]
        
        # Calculate the correlation matrix for the filtered DataFrame
        correlation_matrix = activity_df[columns_of_interest].corr()
        
        # Create a mask for correlations that are below the threshold
        mask = np.triu(np.ones_like(correlation_matrix, dtype=bool)) | (correlation_matrix <= threshold)
        
        # Plotting the correlation matrix using seaborn
        plt.figure(figsize=(12, 10))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', mask=mask)
        plt.title(f'Highly Positive Correlation Matrix for Activity {activity_num}')
        
        # Saving the plot as an image file
        plt.savefig(f'highly_positive_correlation_matrix_activity_{activity_num}.png')
        plt.close()  # Close the plot to avoid displaying it inline

# Call the function to plot and save the highly positive correlation matrix images
plot_highly_positive_correlations(df, columns_of_interest)
```
